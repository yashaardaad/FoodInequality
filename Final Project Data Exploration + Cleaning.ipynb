```python
# Now let's read in the food access data and inspect it 

food_access_data = pd.read_csv(
    "Food Access Research Atlas.csv",
    engine="python",
    quoting=3,
    on_bad_lines="skip",
    encoding_errors="ignore"
)
```

```python
# We can try to fill the null columns 

numeric_cols = food_access_data.select_dtypes(include="number").columns

food_access_data[numeric_cols] = (
    food_access_data[numeric_cols]
    .fillna(0)
)
```

```python
food_access_data.head(17)
```

```python
# To be able to join with the other data, we'll need to drop the word county from all of the different rows when referring to this 

food_access_data["County"] = (
    food_access_data["County"]
    .str.replace(" County", "", regex=False)
    .str.strip()
)
```

```python
food_access_data.head()
```

```python
# Now we can group by and also make sure we include the exact columns in the group by, making sure they're taking the values we want as well 

food_access_county = (
    food_access_data
    .groupby(["State", "County"], as_index=False)
    .agg({
        "Urban": "mean",
        "Low Income": "mean",
        "Low Income and Low Access @ 1 and 10 Miles from Supermarket": "mean",
        "Low Access @ 1 Mile Urban and 10 Miles Rural": "mean",
        "Low Access at 1 Mile": "mean",
        "Low Acess at 10 Miles": "mean",

        **{
            col: "sum"
            for col in food_access_data.columns
            if col not in [
                "State",
                "County",
                "Urban",
                "Low Income",
                "Low Income and Low Access @ 1 and 10 Miles from Supermarket",
                "Low Access @ 1 Mile Urban and 10 Miles Rural",
                "Low Access at 1 Mile",
                "Low Acess at 10 Miles"
            ]
        }
    })
)
```

```python
food_access_county.head()
```

```python
# We can make these values more meaningful by taking into account their weighted averages

food_access_county = (
    food_access_data
    .groupby(["State", "County"])
    .apply(
        lambda x: pd.Series({
            "Population": x["Population"].sum(),

            "Poverty Rate": (
                (x["Poverty Rate"] * x["Population"]).sum()
                / x["Population"].sum()
            ),

            "Share Population Beyond 1 Mile": (
                (x["Share Population Beyond 1 Mile"] * x["Population"]).sum()
                / x["Population"].sum()
            ),

            "Share of Low Income Population Beyond 1 Mile": (
                (x["Share of Low Income Population Beyond 1 Mile"] * x["Population"]).sum()
                / x["Population"].sum()
            ),

            "Share of Houses w/out Vehicle Beyond 1 Mile": (
                (x["Share of Houses w/out Vehicle Beyond 1 Mile"] * x["Population"]).sum()
                / x["Population"].sum()
            ),

            "Share of Houses Receiving SNAP Beyond 1 Mile": (
                (x["Share of Houses Receiving SNAP Beyond 1 Mile"] * x["Population"]).sum()
                / x["Population"].sum()
            )
        })
    )
    .reset_index()
)
```

```python
def _norm(s: pd.Series) -> pd.Series:
    return (
        s.astype(str)
         .str.strip()
         .str.lower()
         .str.replace(r"\s+(county|parish|borough|census area|municipality)$", "", regex=True)
    )

city_level_data = city_level_data.copy()
food_access_county = food_access_county.copy()

city_level_data["_state_key"]  = _norm(city_level_data["State"])
city_level_data["_county_key"] = _norm(city_level_data["County"])
food_access_county["_state_key"]  = _norm(food_access_county["State"])
food_access_county["_county_key"] = _norm(food_access_county["County"])
```

```python
food_cols = [
    "_state_key", "_county_key",
    "Poverty Rate",
    "Share Population Beyond 1 Mile",
    "Share of Low Income Population Beyond 1 Mile",
    "Share of Houses w/out Vehicle Beyond 1 Mile",
    "Share of Houses Receiving SNAP Beyond 1 Mile",
]
```

```python
# We can move to merging the datasets now that we have the proper columns and everything normalized


merged = city_level_data.merge(
    food_access_county[food_cols],
    on=["_state_key", "_county_key"],
    how="left",
    validate="many_to_one",
)

merged = merged.drop(columns=["_state_key", "_county_key"])
```

```python
unmatched = merged["Poverty Rate"].isna().sum()
total     = len(merged)
print(f"Rows: {total:,}  |  Unmatched on food access: {unmatched:,} ({unmatched/total:.1%})")
print(merged.head())
```
